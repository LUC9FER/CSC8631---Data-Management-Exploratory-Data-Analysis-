### Event Run Time

#### Data For Render

```python
Render = Merged[Merged['eventName_x'] == 'Render'] [Merged['eventName_y'] == 'Render'] # Extracting Data As Per 'Render'

Grouped_Render = Render.groupby(['hostname_x']).mean() # Grouping Data According To 'hostname'

Time_Render = Grouped_Render['Time_In_Sec'] # Extracting Execution Time For Render
```

#### Data For Saving Configuration

```python
Saving_Config = Merged[Merged['eventName_x'] == 'Saving Config'] [Merged['eventName_y'] == 'Saving Config'] # Extracting Data As Per 'Saving Config'

Grouped_Saving_Config = Saving_Config.groupby(['hostname_x']).mean() # Grouping Data According To 'hostname'

Time_Saving_Config = Grouped_Saving_Config['Time_In_Sec'] # Extracting Execution Time For Saving Config
```

#### Data For Tiling

```python
Tiling = Merged[Merged['eventName_x'] == 'Tiling'] [Merged['eventName_y'] == 'Tiling'] # Extracting Data As Per 'Tiling'

Grouped_Tiling= Tiling.groupby(['hostname_x']).mean() # Grouping Data According To 'hostname'

Time_Tiling = Grouped_Tiling['Time_In_Sec'] # Extracting Execution Time For Tiling
```

#### Data For Uploading

```python
Uploading = Merged[Merged['eventName_x'] == 'Uploading'] [Merged['eventName_y'] == 'Uploading'] # Extracting Data As Per 'Uploading'

Grouped_Uploading = Uploading.groupby(['hostname_x']).mean() # Grouping Data According To 'hostname'

Time_Uploading = Grouped_Uploading['Time_In_Sec'] # Extracting Execution Time For Tiling
```

#### Visualizing The Results

```python
X = ['Render', 'Saving Config', 'Tiling', 'Uploading'] # Labels For X-Axis


# Calculating Total Time For Every Event

Total_Render = Time_Render.sum()

Total_Saving_Config = Time_Saving_Config.sum() 

Total_Tiling = Time_Tiling.sum()

Total_Uploading = Time_Uploading.sum()

 
Heights = Total_Render, Total_Saving_Config, Total_Tiling, Total_Uploading # Height Of The Bar
```

```python
plt.bar(X, Heights, color=['purple', 'black', 'green', 'blue', 'cyan']) # Plotting Bar Graph

plt.title('Run Time Taken By Each Event') # Giving Title To The Graph

plt.xlabel('Event') # Giving Title To X-Axis

plt.ylabel('Time (In Sec)') # Giving Title To Y-Axis

plt.show() # Showing The Graph
```
