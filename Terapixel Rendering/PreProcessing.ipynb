#                                    Data Pre-Processing

### Importing Libraries

```python
import numpy as np # Library For Arrays

# Library For Graphs

import matplotlib.pyplot as plt 

import seaborn as sns

import pandas as pd # Library For Datasets

import datetime # Library For Manipulating 'datetime'

from datetime import timedelta # Importing 'timedelta' Class To Calculate Seconds
```

### Importing Dataset

```python
Application_CheckPoints = pd.read_csv('../data/application-checkpoints.csv') # Importing 'application-checkpoints' Dataset 

GPU = pd.read_csv('../data/gpu.csv') # Importing 'gpu' Dataset

Application_CheckPoints.head() # Printing First Few Results
```

### Changing Data-Time Format 

```python
Old_Date_Time = Application_CheckPoints['timestamp'] # Extracting Date-Time Data From The 'Application_CheckPoints' Dataset

# Changing The Date-Time Format & Calculating Time In Seconds

def date_convert(Old_Date_Time):
    
    Date_Time = datetime.datetime.strptime(Old_Date_Time, '%Y-%m-%dT%H:%M:%S.%fZ')
    return Date_Time

    
Application_CheckPoints['timestamp'] = Application_CheckPoints['timestamp'].apply(date_convert) # Applying The Changes
```

```python
Application_CheckPoints.head() # Printing Few Results
```

### Merging Datasets

####  * Grouping GPU Data According To 'hostname'

```python
Grouped_GPU = GPU.groupby(['hostname']).mean() # Grouping The Data According To 'hostname'
```

#### *  Merging Grouped GPU & Merged Application CheckPoints Data According To 'hostname'

```python
Application_GPU_Merged = pd.merge(Grouped_GPU, Application_CheckPoints, on = ['hostname'], how = 'left')

Application_GPU_Merged.head() # Printing The First Few Results
```

### Sorting Data According To Task ID

```python
Start_Data = Application_GPU_Merged[Application_GPU_Merged['eventType'] == 'START'] # Extracting Data Of 'START'

Stop_Data = Application_GPU_Merged[Application_GPU_Merged['eventType'] == 'STOP'] # Extracting Data Of 'STOP'

Merged = pd.merge(Start_Data, Stop_Data, on = 'taskId', how = 'left') # Merging 'START' & 'STOP' Data As Per Task ID 

Merged.head() # Printing First Few Results
```

### Calculating Time Difference

```python
Execution = Merged['timestamp_y'] - Merged['timestamp_x'] # Calculating Difference Between 'START' & 'STOP' Time

Seconds = Execution.dt.total_seconds() # Extracting Seconds

Merged['Execution_Time'] = Execution # Creating A New Column & Storing Values Of Difference

Merged['Time_In_Sec'] = Seconds # Storing Seconds

Merged.head() # Printing First Few Data
```

### Exporting The File

```python
Merged.to_csv('../data/Merged.csv') # Exporting In '.csv' Form
```
